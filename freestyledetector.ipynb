## Segundo paso

```python
import cv2
from deepface import DeepFace

# Inicializar cámara
cap = cv2.VideoCapture(0)

def analizar_y_reaccionar():
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        try:
            # Analizar el frame con DeepFace
            resultados = DeepFace.analyze(
                frame, 
                actions=['emotion', 'age', 'gender'],
                detector_backend="opencv",
                enforce_detection=False
            )
            
            if len(resultados) > 0:
                emocion = resultados[0]['dominant_emotion']
                edad = resultados[0]['age']
                
                # TU LÓGICA CREATIVA AQUÍ
                if emocion == "happy":
                    cv2.putText(frame, "¡SONRIE! ", (50, 50), 
                               cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 3)
                    # Dibujar elementos creativos
                    cv2.circle(frame, (100, 120), 30, (0, 255, 255), -1)  # Sol
                    
                elif emocion == "sad":
                    cv2.putText(frame, "ANIMO! ", (50, 50), 
                               cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255, 0, 0), 3)
                    # Dibujar lluvia
                    for i in range(5):
                        cv2.line(frame, (80 + i*30, 100), (80 + i*30, 130), 
                                (255, 255, 255), 2)
                
                # Mostrar información
                cv2.putText(frame, f"Edad: {edad}", (50, 400), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
                cv2.putText(frame, f"Emoción: {emocion}", (50, 430), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
                
        except Exception as e:
            print(f"Error: {e}")
        
        cv2.imshow('Prototipo Creativo - Reacciones Emocionales', frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()

# Ejecutar
analizar_y_reaccionar()
```
